```python
import cv2
import torch
import numpy as np
import os 
os.chdir(r"E:\Download")
area = []

def POINTS(event, x, y, flags, param):
    if event == cv2.cv2.EVENT_LBUTTONDOWN :  
        point_xy = [x, y]
        
        print(point_xy)
        
area=[[26, 228], [12, 275], [439, 225], [394, 190]] 
cv2.namedWindow('FRAME')
           
model = torch.hub.load('ultralytics/yolov5', 'yolov5s', pretrained=True)
cap=cv2.VideoCapture('parking.mp4')
while True:
    
    ret,frame=cap.read()
    if not ret:
        break
       
    cv2.imshow("FRAME",frame)
    
    
    results=model(frame)
    print(results.pandas().xyxy[0])
    points = []
    for index, row in results.pandas().xyxy[0].iterrows():
        if row["name"]== "car":
            x1 = int(row['xmin'])
            y1 = int(row['ymin'])
            x2 = int(row['xmax'])
            y2 = int(row['ymax'])
        
            cx=int(x1+x2)//2
            cy=int(y1+y2)//2
            results = cv2.pointPolygonTest(np.array(area,np.int32),(cx,cy),False)
            if results>=0:
                cv2.rectangle(frame,(x1,y1),(x2,y2),(0,0,255),3)
                cv2.putText(frame,"car",(x1,y1),cv2.FONT_HERSHEY_PLAIN,2,(255,0,0),2)
                points.append([cx])
    a = len(points)
    cv2.polylines(frame,[np.array(area,np.int32)],True,(0,255,0),2)
    cv2.putText(frame,'number of cars in parking ='+str(a),(50,80),cv2.FONT_HERSHEY_PLAIN,2,(0,0,255),2)
    cv2.imshow("FRAME",frame)
    cv2.setMouseCallback("FRAME",POINTS)
    if cv2.waitKey(1)==27:
        break
cap.release()


cv2.destroyAllWindows()

```
